```python
# Первый вариант rename
df = df.rename(columns={'Hotel':'hotel', 'Is Canceled':'is_canceled', 'Lead Time':'lead_time', 'arrival full date':'arrival_full_date',
       'Arrival Date Year':'arrival_date_year', 'Arrival Date Month':'arrival_date_month', 'Arrival Date Week Number':'arrival_date_week_number',
       'Arrival Date Day of Month':'arrival_date_day_of_month', 'Stays in Weekend nights':'stays_in_weekend_nights',
       'Stays in week nights':'stays_in_week_nights', 'stays total nights':'stays_total_nights', 'Adults':'adults', 'Children':'children',
       'Babies':'babies', 'Meal':'meal', 'Country':'country', 'Reserved Room Type':'reserved_room_type', 'Assigned room type':'assigned_room_type',
       'customer type':'customer_type', 'Reservation Status':'reservation_status', 'Reservation status_date':'reservation_status_date'})
```

```python
df
```

```python
# Топ 5 стран по бронированию отелей
# При 0 бронь остается, при 1 слетает
top_5 = df.query('is_canceled == 0').groupby('country',as_index=False)['is_canceled'].agg('count').sort_values('is_canceled',ascending=False).head(5)
```

```python
top_5
```

```python
# Среднее кол-во дней, на которые бронировали отели
avg_hotel = df.groupby('hotel', as_index=False)['stays_total_nights'].agg('mean')
```

```python
avg_hotel
```

```python
# Тип номера, полученного клиентом (assigned_room_type), отличается от изначально забронированного
reversed_error = df.query('assigned_room_type != reserved_room_type').agg('count')
```

```python
reversed_error
```

```python
# Из даты сета видно, что в 2016 году октябрь является самым посещаемым месяцев в году
most_popular_2016 = df.query('arrival_date_year == 2016').groupby('arrival_date_month',as_index=False)['arrival_date_year'].agg('count').sort_values('arrival_date_year', ascending=False)
```

```python
most_popular_2016
```

```python
most_popular_2017 = df.query('arrival_date_year == 2017').groupby('arrival_date_month',as_index=False)['arrival_date_year'].agg('count').sort_values('arrival_date_year', ascending=False)
```

```python
most_popular_2017
```

```python
most_canceled = df.query('is_canceled == 1 and hotel == "City Hotel"').groupby(['arrival_date_year','arrival_date_month'], as_index=False)['is_canceled'].agg('count')
```

```python
# в 2015 в сентябре, в октябре в 2016, в мае 2017 - самые большие отмены
most_canceled
```

```python
df['total_kids'] = df['children'] + df['babies']
```

```python
most_mean_kids = df.groupby('hotel', as_index=False)['total_kids'].agg('mean')
```

```python
df['has_kids'] = np.where(df['total_kids'] > 0, True, False)
```

```python
df
```

```python
# Churn rate
churn_rate = (df[df['is_canceled'] == 1].count()) / (df['is_canceled'].count())
```

```python
churn_rate = churn_rate['hotel']
```

```python
churn_rate
```

```python
cr_hasnt_kids = df.query('is_canceled == 1 and has_kids == False').shape[0] / df.query('has_kids == False').shape[0]
```

```python
cr_hasnt_kids = round(cr_hasnt_kids, 2) * 100
```
